```python
X = df[['CRIM', 'INDUS ', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'LSTAT']]
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_test,predictions)
RSquare = metrics.r2_score(y_test,predictions)
MSE = metrics.mean_squared_error(y_test,predictions)
RMSE = np.sqrt(MSE)
```

```python
print('The MAE of the model is: ', MAE)
print('The MSE of the model is: ', MSE)
print('The RMSE of the model is: ', RMSE)
print('The RSquare of the model is: ', RSquare)
```
